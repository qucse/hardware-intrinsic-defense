```python
import os
import pickle
import numpy as np
import torch
from datasets import load_dataset
from torchvision import transforms
from PIL import Image
from torch.utils.data import Dataset, DataLoader
```

```python


def create_miniimagenet_batches(output_dir, batch_size):
    """
    Loads Mini-ImageNet from Hugging Face and saves it in structured batches.
    Each batch file contains image file paths and labels, without resizing images.
    
    - output_dir: Destination for batch files
    - batch_size: Number of images per batch
    """
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    print("🔄 Loading Mini-ImageNet from Hugging Face...")
    dataset = load_dataset("timm/mini-imagenet", split="train")

    # Get unique class names and create a mapping
    classes = sorted(set(example["label"] for example in dataset))
    class_to_idx = {cls: idx for idx, cls in enumerate(classes)}

    # Extract image file paths & labels
    image_data = []
    for example in dataset:
        image_data.append((example["image"], class_to_idx[example["label"]]))

    # Shuffle dataset
    np.random.shuffle(image_data)

    # Save metadata (class names)
    meta = {"label_names": classes}
    with open(os.path.join(output_dir, "batches.meta"), "wb") as f:
        pickle.dump(meta, f)

    batch_num = 1
    data_batch = []
    labels_batch = []

    for image, label in image_data:
        data_batch.append(image)  # Store PIL image directly
        labels_batch.append(label)

        if len(data_batch) == batch_size:
            batch_dict = {"images": data_batch, "labels": labels_batch}
            batch_filename = os.path.join(output_dir, f"data_batch_{batch_num}.pkl")
            with open(batch_filename, "wb") as f:
                pickle.dump(batch_dict, f)

            print(f"✅ Saved batch {batch_num} with {batch_size} images.")
            batch_num += 1
            data_batch = []
            labels_batch = []

    # Save remaining images
    if data_batch:
        batch_dict = {"images": data_batch, "labels": labels_batch}
        batch_filename = os.path.join(output_dir, f"data_batch_{batch_num}.pkl")
        with open(batch_filename, "wb") as f:
            pickle.dump(batch_dict, f)
        print(f"✅ Saved batch {batch_num} with {len(data_batch)} images.")

    print("🎉 Mini-ImageNet dataset is now structured into batches!")

if __name__ == "__main__":
    output_dir = r"D:\AIoT_project\data_sets\miniimagenet_batches"
    batch_size = int(input("Enter batch size (e.g., 10,000 images per batch): "))
    create_miniimagenet_batches(output_dir, batch_size)
```
